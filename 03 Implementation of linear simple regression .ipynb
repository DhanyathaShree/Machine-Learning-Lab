### 3. Implementation of linear simple regression 

```python
import pandas as pd
import matplotlib.pyplot as plt
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]
n=len(x)
xsum=sum(x)
ysum=sum(y)
xysum=0
x2sum=0
for i in range(n):
    b=x[i]*y[i]
    xysum+=b
    x2sum+= x[i]**2
print(xsum,ysum,xysum,x2sum)
#calculate slope and intercept
m= (n*xysum-xsum*ysum)/(n*x2sum - xsum**2)
c=(ysum-m*xsum)/n
print(f'y = {m}x + {c}')

data = {'x': x, 'y': y}
df = pd.DataFrame(data)
df['y_predict'] = df['x'].apply(lambda i: m * i + c)

plt.scatter(df.x,df.y,color='b',label='y points')
plt.plot(df.x,df.y_predict,color='r',marker='o',label='regression line')
plt.legend()
plt.show()
print("\nName : Dhanyatha Shree C A\nRoll No : CB.SC.I5DAS21023\n")
```
