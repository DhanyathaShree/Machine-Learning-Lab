```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn import datasets 
from sklearn.naive_bayes import GaussianNB 
from sklearn.preprocessing import *

df=pd.read_csv('naivebayes.csv')
x_=df.iloc[:,:-1]
y_=df['Play Golf']
df
```

```python
from math import *

df=pd.read_csv('naivebayes.csv')
x_=df.iloc[:,:-1]
data = {'values':{},'yes':{},'no':{},'p(yes)':{},'p(no)':{}}
table = pd.DataFrame(data)

c=x_.columns.tolist()
for x in range(len(c)):
    coll=df[c[x]].unique().tolist()
    yes,no=[],[]
    for i in coll:
        y=len(df.loc[(df[c[x]]==i)&(df['Play Golf']=='Yes')])
        n=len(df.loc[(df[c[x]]==i)&(df['Play Golf']=='No')])
        yes.append(y)
        no.append(n)
    pyes = [q / sum(yes) for q in yes]
    pno = [q / sum(no) for q in no]
    data = {'values': coll,'yes': yes,'no':no,'p(yes)':pyes,'p(no)':pno}
    tablev = pd.DataFrame(data)
    table = pd.concat([table, tablev], ignore_index=True)
    
table
```

```python
listy=list(y_)
df_yes = listy.count('Yes')/len(df)
df_no = listy.count('No')/len(df) 

input=['Rainy','Hot','High','True']
yes_prob,no_prob=0,0

input=['Rainy','Hot','High','True']

y=table.loc[table['values'].isin(input)]
yes_prob= prod(y['p(yes)'].tolist()) * df_yes
no_prob= prod(y['p(no)'].tolist()) * df_no
tot_prob= yes_prob+no_prob
yes_possibility= yes_prob/tot_prob
no_possibility= no_prob/tot_prob
print(f'p(yes) = {yes_possibility} , p(no) = {no_possibility}')
print()
if yes_possibility>no_possibility:
    print("Yes")
else:
    print("No")

print("\nName : Dhanyatha Shree C A\nRoll No : CB.SC.I5DAS21023\n")
```
